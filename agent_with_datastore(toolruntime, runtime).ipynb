```python
from dataclasses import dataclass
from langchain.agents import create_agent
from langchain.tools import tool, ToolRuntime
from langgraph.runtime import Runtime
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model
from langgraph.store.memory import InMemoryStore
load_dotenv()
```

```python
store = InMemoryStore()
llm = init_chat_model(
    model = "gpt-4o-mini",
    model_provider='openai'
)
```

```python
@dataclass
class Context:
    user_name: str


@tool
def fetch_user_name(runtime: ToolRuntime[Context]) -> str:  
    """Fetch the user's name from the store."""
    user_name = runtime.context.user_name
    return user_name

@tool
def add(a: float, b: float)->float:
    """
    Performs addition of two number
    """
    return a+b



agent_prompt = """You are an addition operation performing agent, Evaluate the user's expression
Ensure you always respond user with his name"""

agent = create_agent(
    model="gpt-4o-mini",
    tools=[fetch_user_name, add],
    context_schema=Context,
    system_prompt=agent_prompt,
)

agent.invoke(
    {"messages": [{"role": "user", "content": "What's 2+3+6?"}]},
    context=Context(user_name="Narasimhulu")  
)
```
